# Data Imputation

```python
df_fill = df.copy(deep=True)
df_fill = df_fill.fillna(160)
df.describe()
```

```python
df.describe()
```

```python
df_fill.count()
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer

SI = SimpleImputer(missing_values=np.nan, strategy='mean')
SI.fit(df[['Time_taken']])
df['Time_taken'] = SI.transform(df[['Time_taken']])
```

```python
df.isna().sum()
```
